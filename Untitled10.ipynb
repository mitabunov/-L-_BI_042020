```python


class RecognitionObject:
    def __init__(self, image_path, what_looking):
        self.image_path = image_path
        self.what_looking = what_looking

    def recognition_image(self):
        self.model = tf.keras.applications.EfficientNetV2B0(weights='imagenet')
        self.image = cv2.imread(self.image_path)
        image = cv2.cvtColor(self.image, cv2.COLOR_BGR2RGB)
        image = tf.image.resize(image, (224,224))
        image = tf.expand_dims(image, axis=0)

        predictions = self.model.predict(image)
        decoded_predictions = tf.keras.applications.imagenet_utils.decode_predictions(predictions)[0]
        try:
            for i, (imagenet_id, label, score) in enumerate(decoded_predictions):
                print(f'{i+1}:{label} ({score: 2f})')
                if what_looking in label.lower():
                    h, w, _ = image.shape
                    cv2.putText(image, f'{label} ({score: 2f})',(10,30), cv2.FONT_HERSHEY_TRIPLEX)
                    cv2.rectangle(image, (0,0), (w, h), (244,0,0), 2)

            image = cv2.cvtColor(image.numpy()[0], cv2.COLOR_BGR2RGB)
            # cv2.imshow('Object', image)
            cv2_imshow(image)

            cv2.waitKey(0)
            cv2.destroyAllWindows()

        except AttributeError:
          print()
```
